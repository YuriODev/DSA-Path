## <a id='toc1_3_'></a>[Theory](#toc0_)

### <a id='toc1_3_1_'></a>[What is an Algorithm? ü§ñ](#toc0_)

**Definition:** An algorithm is a finite, step-by-step procedure for solving a problem or performing a task.

**Key Characteristics:**
- **Input/Output:** Accepts input and produces output.
- **Definiteness:** Each step is clearly defined with no ambiguity.
- **Finiteness:** The process must conclude after a finite number of steps.
- **Effectiveness:** Each step must be simple enough to be executed, typically by a computer.

**Deep Dive:**
Algorithms form the backbone of all computing processes. They are used not only in computer science but also in everyday tasks. For example, a cooking recipe is an algorithm: you have ingredients (input) and a delicious dish (output) üç∞.

Designing an algorithm involves breaking down a problem into smaller, manageable parts and then solving each part in a logical sequence. This systematic approach boosts problem-solving skills and ensures efficiency. üöÄ

### <a id='toc1_3_2_'></a>[Introduction to Time Complexity ‚è±Ô∏è](#toc0_)

**Definition:** Time complexity describes the amount of time an algorithm takes to run as a function of the input size.

**Big-O Notation:**
Big-O notation provides an upper bound on the growth rate of an algorithm's running time. It allows us to compare the efficiency of different algorithms at a glance.

**Common Time Complexities:**
- **O(1) ‚Äì Constant Time:** Execution time remains constant regardless of input size. ‚ö°
- **O(log n) ‚Äì Logarithmic Time:** Execution time increases logarithmically as input size increases. üîç
- **O(n) ‚Äì Linear Time:** Execution time grows linearly with the input size. üìà
- **O(n¬≤) ‚Äì Quadratic Time:** Execution time grows proportionally to the square of the input size. üìä

**Additional Considerations:**
- **Best-case, Average-case, Worst-case:** Different scenarios can influence performance analysis.
- **Amortized Analysis:** Sometimes, a costly operation is rare and is offset by many cheaper ones.

Understanding time complexity is crucial when processing large datasets, as inefficient algorithms can lead to significant delays and resource usage. üî•

## <a id='toc1_5_'></a>[Code Examples & Visualizations üíª](#toc0_)

### <a id='toc1_5_1_'></a>[Code Example: Summing a List (Linear Algorithm) üìù](#toc0_)

In this example, we define a function that sums the elements of a list using a loop. This demonstrates a linear time (O(n)) approach. üîÑ

````python
def sum_list(numbers):
    total = 0
    for num in numbers:
        total += num
    return total
````

The function `sum_list` iterates over each element in the list and accumulates the sum. Since the loop runs once per element, the time complexity is **O(n)**. ‚è±Ô∏è

````python
data = list(range(10000))
print("Sum:", sum_list(data))
````

The output `Sum: 49995000` confirms that the `sum_list` function correctly sums numbers from 0 to 9999.

Now, let's measure the execution time of `sum_list` using Jupyter's `%timeit` magic command:

````python
%timeit sum_list(data)
````

The `%timeit` output shows the average execution time for the custom `sum_list` function over multiple runs.

### <a id='toc1_5_2_'></a>[Code Example: Using Python's Built-in `sum()` Function ‚ö°](#toc0_)

````python
result = sum(data)
print("Built-in Sum:", result)
````

The built-in `sum()` function returns the same result (`49995000`), but is more optimized due to its underlying C implementation.

Measure the execution time of the built-in `sum()` function:

````python
%timeit sum(data)
````

The `%timeit` result for the built-in `sum()` function is significantly lower than that of the custom function, demonstrating its superior performance.

### <a id='toc1_5_3_'></a>[Visualization: Plotting Runtime vs. Input Size üìä](#toc0_)

````python
import matplotlib.pyplot as plt
import numpy as np
import time

def linear_time_function(n):
    start = time.time()
    total = 0
    for i in range(n):
        total += i
    return time.time() - start

# Generate sample input sizes
sizes = np.linspace(1000, 10000, 10, dtype=int)
times = [linear_time_function(n) for n in sizes]

display("Measured Times:", times)
````

````python
plt.figure(figsize=(10, 6))
plt.plot(sizes, times, linestyle='-', color='teal', marker='o', label='O(n) - Linear')
plt.xlabel("Input Size (n)")
plt.ylabel("Time (seconds)")
plt.title("Runtime vs. Input Size for a Linear Algorithm")
plt.legend()
plt.grid(True)
plt.show()
````

The graph shows that the runtime increases linearly with the input size, confirming the **O(n)** time complexity of the algorithm.

### <a id='toc1_5_4_'></a>[Comparative Graph: Theoretical Time Complexities üåà](#toc0_)

````python
x = np.linspace(1, 40, 40)
constant = np.ones_like(x) * 5  # constant value scaled for visualization
logarithmic = np.log(x) * 10
linear = x
quadratic = (x ** 2) / 20

plt.figure(figsize=(10, 6))
plt.plot(x, constant, label='O(1) - Constant', linestyle='-', color='red')
plt.plot(x, logarithmic, label='O(log n) - Logarithmic', linestyle='--', color='green')
plt.plot(x, linear, label='O(n) - Linear', linestyle='-.', color='blue')
plt.plot(x, quadratic, label='O(n¬≤) - Quadratic', linestyle=':', color='purple')

plt.xlabel("Input Size (n)")
plt.ylabel("Operations (normalized)")
plt.title("Comparison of Time Complexities")
plt.legend()
plt.grid(True)
plt.show()
````

The comparative graph illustrates the growth rates of different time complexities, showing that:
- **O(1)** remains constant,
- **O(log n)** grows very slowly,
- **O(n)** grows linearly, and
- **O(n¬≤)** grows quadratically with input size.
